## 1.1 The Limits of Traditional Machine Learning

In traditional ML (e.g. SVMs, Random Forests, Logistic Regression), **you** must design the features fed into the model.

For simple tabular data (age, salary, height), this works fine.  
But for **images, audio, and text**, hand-crafting features becomes:
- Extremely difficult
- Time-consuming
- Often suboptimal

> 🧠 **Key Insight:** Deep learning **learns the features automatically** from raw data.

```
Traditional ML pipeline:
  Raw Data → [Hand-crafted Features] → ML Model → Prediction

Deep Learning pipeline:
  Raw Data → [Neural Network learns features AND predicts] → Prediction
```


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_moons

# Traditional ML: Logistic Regression on a non-linear problem
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)

model = LogisticRegression()
model.fit(X, y)

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(-2, 3, 200), np.linspace(-1.5, 2, 200))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8, 4))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='RdBu')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdBu', edgecolors='k', s=40)
plt.title("Logistic Regression on Moon Dataset\n(Linear model struggles with non-linear data)")
plt.xlabel("Feature 1"); plt.ylabel("Feature 2")
plt.tight_layout()
plt.show()
print(f"Accuracy: {model.score(X, y):.2f}")
````

## 1.2 Hierarchical Feature Learning

Deep networks learn **layers of abstractions**:

```
Image Input
    ↓
Layer 1: Detects edges and corners
    ↓
Layer 2: Detects shapes (eyes, wheels, doors)
    ↓
Layer 3: Detects objects (faces, cars)
    ↓
Output: "This is a cat"
```

Each layer builds on top of the previous one.  
This is called **representation learning** or **feature learning**.


## What is “deep” learning?

It’s a fuzzy term for **artificial neural networks** with “many” hidden layers.

## 1.3 The Universal Approximation Theorem (Intuition)

> 📐  A neural network with even **one hidden layer** and enough neurons can approximate *any* continuous function.

This means neural networks are **theoretically capable** of learning any pattern.

In practice, **depth** (more layers) is more efficient than **width** (more neurons per layer) for complex tasks.


## 1.4 When to Use Deep Learning vs Traditional ML

| Situation | Use |
|---|---|
| Tabular dataset | Traditional ML (Random Forest, XGBoost) |
| Large dataset, raw images | Deep Learning (CNN) |
| Text / NLP tasks | Deep Learning (Transformers) |
| Audio / Speech | Deep Learning (RNN, CNN) |
| Need interpretability | Traditional ML |
| Limited compute | Traditional ML |

> 💡 Deep learning shines when you have **lots of data** and **raw, unstructured inputs**.
