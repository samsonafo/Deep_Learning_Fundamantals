# 6. The Chain Rule — The Heart of Backprop
---
Before we can understand backpropagation, we need to understand the **chain rule** from calculus. This is the single mathematical concept that makes training deep networks possible.

The instantaneous rate of change of a function with respect to one of its variables. It measures how quickly a function is changing at a specific point, often visualized as the slope of the tangent line to a curve. (Google)


## 6.1 Derivatives: A Quick Refresher

The **derivative** $\frac{df}{dx}$ tells you: *if I nudge $x$ by a tiny amount, how much does $f(x)$ change?*

$$\frac{df}{dx} \approx \frac{f(x + \epsilon) - f(x)}{\epsilon} \quad \text{(for very small } \epsilon\text{)}$$

In deep learning, we need to know: *if I nudge a weight $w$ by a tiny amount, how much does the **loss** change?*

That's the gradient $\frac{\partial L}{\partial w}$ — and the chain rule is how we compute it.

## 6.2 The Chain Rule

If $y = g(x)$ and $z = f(y) = f(g(x))$, then:

$$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}$$

In English: *to find how $z$ changes with $x$, multiply how $z$ changes with $y$ by how $y$ changes with $x$.*

**Example:**  
Let $y = 2x + 1$ and $z = y^2$:
- $\frac{dy}{dx} = 2$
- $\frac{dz}{dy} = 2y$
- $\frac{dz}{dx} = 2y \cdot 2 = 4y = 4(2x+1)$


## 6.3 Chain Rule for Longer Chains

For a chain $x → y → z → w$:

$$\frac{dw}{dx} = \frac{dw}{dz} \cdot \frac{dz}{dy} \cdot \frac{dy}{dx}$$

This generalizes to any depth! A neural network with 100 layers just has a chain of 100 terms.

**Backpropagation is literally just applying the chain rule efficiently**, going from the output back to the inputs.


## 6.4 Computational Graphs

It helps to visualize computation as a **graph**, where:
- Each **node** is an operation (+, ×, ReLU, σ, ...)
- Each **edge** carries a value (or gradient during backprop)

```
Forward (left → right):
  x ─→ [×3 +1] ─→ y ─→ [ReLU] ─→ z ─→ [σ] ─→ w ─→ [(·-1)²] ─→ Loss

Backward (right → left, applying chain rule at each node):
  ∂L/∂x ←─ [×3] ←─ ∂L/∂y ←─ [×ReLU'] ←─ ∂L/∂z ←─ [×σ'] ←─ ∂L/∂w ←─ [×2(w-1)] ←─ 1.0
```

Each backward step **multiplies** by the local gradient of that node.
