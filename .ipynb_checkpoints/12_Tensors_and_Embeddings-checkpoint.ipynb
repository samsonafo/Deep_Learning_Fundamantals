# 12. Tensors & Embeddings
---

How do we feed **text, categories, or discrete tokens** into a neural network? The answer is **embeddings** — dense vector representations learned during training.


## 12.1 The Problem: Discrete Inputs

Neural networks need **continuous numbers** as input. But many real-world inputs are discrete:
- Words: "cat", "dog", "Berlin"
- Categories: city names, product IDs, user IDs
- Tokens in a language model

**Naive approach — One-hot encoding:**
For a vocabulary of 50,000 words → each word is a vector of 50,000 zeros with one 1.

Problems:
- 🗂️ Huge memory: 50,000 × 50,000 matrix
- 🙈 No meaning: "king" and "queen" are as different as "king" and "chair"
- 📉 Sparse: almost all zeros, no gradient signal

**Solution: Embeddings** — map each token to a **dense low-dimensional vector** that is *learned during training*.


````python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

torch.manual_seed(42)

# One-hot vs embedding comparison
vocab_size = 10
embed_dim  = 3

# One-hot: sparse, 10-dimensional
one_hot = torch.eye(vocab_size)
print("One-hot encoding (vocab=10):")
print(one_hot)
print(f"Shape: {one_hot.shape}  — {one_hot.numel()} values, {(one_hot==0).sum().item()} are zeros")
````

````python
# Embedding: dense, low-dimensional
embedding = nn.Embedding(num_embeddings=vocab_size, embedding_dim=embed_dim)

# Lookup: just pass integer token IDs
token_ids = torch.tensor([0, 1, 2, 3])   # indices into the vocabulary
vectors = embedding(token_ids)

print("Embedding lookup (vocab=10, dim=3):")
print(vectors)
print(f"\nShape: {vectors.shape}  — each token → {embed_dim}-dim dense vector")
print(f"Parameters: {embedding.weight.numel()} (learned during training)")
print(f"\nEmbedding weight matrix (10×3):")
print(embedding.weight.data.numpy().round(3))
````

## 12.2 How `nn.Embedding` Works

`nn.Embedding(vocab_size, embed_dim)` is essentially a **lookup table**:

```
Embedding table (shape: vocab_size × embed_dim):
  Token 0 → [0.12, -0.45,  0.88, ...]
  Token 1 → [0.67,  0.23, -0.11, ...]
  Token 2 → [-0.34, 0.91,  0.05, ...]
  ...

Input:  token_id = 2
Output: [-0.34, 0.91, 0.05, ...]   ← row 2 of the table
```

Backprop updates only the rows that were used in the forward pass.


````python
# Demonstrate: embedding is just an indexed matrix multiply
vocab_size, embed_dim = 6, 4
emb = nn.Embedding(vocab_size, embed_dim)

token_id = 3
looked_up = emb(torch.tensor(token_id))

# Same result via one-hot @ weight matrix
one_hot_vec = torch.zeros(vocab_size)
one_hot_vec[token_id] = 1.0
manual_lookup = one_hot_vec @ emb.weight

print(f"Token ID: {token_id}")
print(f"nn.Embedding lookup:  {looked_up.detach().numpy().round(4)}")
print(f"one-hot @ W:          {manual_lookup.detach().numpy().round(4)}")
print(f"Match: {'✓' if torch.allclose(looked_up, manual_lookup) else '❌'}")
print(f"\n→ Embedding is a fast indexed version of one_hot @ W")
print(f"→ But much more memory efficient (no need to materialize the one-hot)")
````

## 12.3 Learning Embeddings — A Sentiment Example

Let's train a simple sentiment classifier on movie review sentences.  
The model will **learn embeddings** from scratch to encode meaning.


````python
# Toy sentiment dataset
sentences = [
    "the movie was great",       "i loved the film",
    "absolutely fantastic story","wonderful acting performance",
    "the movie was terrible",    "i hated this film",
    "boring and awful story",    "worst movie ever seen",
    "it was okay i guess",       "not bad not great either",
]
labels = [1,1,1,1,0,0,0,0,2,2]   # 1=positive, 0=negative, 2=neutral

# Build vocabulary
words = sorted(set(w for s in sentences for w in s.split()))
word2idx = {w: i+1 for i, w in enumerate(words)}  # 0 = padding
word2idx['<PAD>'] = 0
vocab_size = len(word2idx) + 1

print(f"Vocabulary size: {len(word2idx)} words")
print(f"Sample: {dict(list(word2idx.items())[:8])}")

# Encode sentences (pad to max length)
max_len = max(len(s.split()) for s in sentences)
encoded = torch.zeros(len(sentences), max_len, dtype=torch.long)
for i, s in enumerate(sentences):
    for j, w in enumerate(s.split()):
        encoded[i, j] = word2idx[w]

y = torch.tensor(labels)
print(f"\nEncoded shape: {encoded.shape}")
print(f"First sentence: {encoded[0]}")
````

````python
class SentimentNet(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, n_classes):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.fc1 = nn.Linear(embed_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, n_classes)

    def forward(self, x):
        # x: (batch, seq_len)
        emb = self.embedding(x)           # (batch, seq_len, embed_dim)
        pooled = emb.mean(dim=1)          # average over words → (batch, embed_dim)
        return self.fc2(self.relu(self.fc1(pooled)))

model = SentimentNet(vocab_size=vocab_size, embed_dim=8, hidden_dim=16, n_classes=3)
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)
criterion = nn.CrossEntropyLoss()

# Train
losses = []
for epoch in range(300):
    optimizer.zero_grad()
    logits = model(encoded)
    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

plt.figure(figsize=(8, 3))
plt.plot(losses, color='#3498DB', lw=2)
plt.title("Sentiment Classifier Training Loss"); plt.xlabel("Epoch"); plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()

preds = model(encoded).argmax(1)
labels_map = {0: 'negative', 1: 'positive', 2: 'neutral'}
print("\nPredictions:")
for s, true, pred in zip(sentences, labels, preds.tolist()):
    status = '✓' if true == pred else '✗'
    print(f"  {status}  '{s[:35]}...' → {labels_map[pred]}")
````

## 12.4 Visualizing Embeddings

After training, similar words should have **similar vectors**.  
We use PCA to project the high-dimensional embeddings into 2D for visualization.


````python
# Extract and visualize learned word embeddings
with torch.no_grad():
    all_ids = torch.tensor([word2idx[w] for w in words])
    all_vecs = model.embedding(all_ids).numpy()

# PCA to 2D
pca = PCA(n_components=2)
vecs_2d = pca.fit_transform(all_vecs)

# Color by sentiment association
pos_words = {'great','loved','fantastic','wonderful','acting'}
neg_words = {'terrible','hated','boring','awful','worst'}

colors = []
for w in words:
    if w in pos_words: colors.append('#2ECC71')
    elif w in neg_words: colors.append('#E74C3C')
    else: colors.append('#95A5A6')

plt.figure(figsize=(10, 7))
plt.scatter(vecs_2d[:, 0], vecs_2d[:, 1], c=colors, s=120, zorder=5, edgecolors='k', lw=0.5)
for i, w in enumerate(words):
    plt.annotate(w, (vecs_2d[i, 0]+0.02, vecs_2d[i, 1]+0.02), fontsize=9)

from matplotlib.patches import Patch
legend_elems = [Patch(facecolor='#2ECC71', label='Positive words'),
                Patch(facecolor='#E74C3C', label='Negative words'),
                Patch(facecolor='#95A5A6', label='Neutral/other')]
plt.legend(handles=legend_elems, loc='best')
plt.title("Learned Word Embeddings (projected to 2D with PCA)
Similar words cluster together!", fontsize=12)
plt.xlabel("PCA Component 1"); plt.ylabel("PCA Component 2")
plt.grid(alpha=0.3); plt.tight_layout(); plt.show()
````

## 12.5 Embeddings for Categories (Not Just Words)

Embeddings work for **any discrete input**, not just words:

```python
# User embeddings (for recommendation systems)
user_embedding = nn.Embedding(num_users=100000, embedding_dim=64)

# Item embeddings
item_embedding = nn.Embedding(num_items=50000, embedding_dim=64)

# City/country embeddings
city_embedding = nn.Embedding(num_cities=5000, embedding_dim=16)

# A recommendation score:
user_vec = user_embedding(user_id)   # (batch, 64)
item_vec = item_embedding(item_id)   # (batch, 64)
score = (user_vec * item_vec).sum(dim=1)   # dot product
```


````python
# Categorical embedding demo: day-of-week embedding for time series
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
day_embedding = nn.Embedding(num_embeddings=7, embedding_dim=4)

# Suppose we have a sequence of days
day_sequence = torch.tensor([0, 1, 2, 3, 4, 5, 6])  # Mon–Sun
day_vecs = day_embedding(day_sequence)

print("Day-of-week embeddings (dim=4):")
for day, vec in zip(days, day_vecs.detach().numpy()):
    print(f"  {day}: {vec.round(3)}")
print()
print("These start random but after training on a task (e.g. sales prediction)")
print("weekday and weekend embeddings would cluster together.")
````
